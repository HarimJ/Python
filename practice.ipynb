```python
import pymysql

class People:
    def __init__(self):
        self.name = ''
        self.sex = ''
        self.age = 0
        self.addr = ''
        
    def inputData(self):
        self.name = input("이름을 입력하시오 : ")
        self.sex = input("성별을 입력하시오(M/W) : ")
        self.age = int(input("나이를 입력하시오 : "))
        self.addr = input("주소를 입력하시오 : ")
        
    def __str__(self):
        return '%5s %3s %3s %7s' % (self.name, self.sex, self.age, self.addr)
    
    def __cmp__(self, other):
        return self.name == other.name
    
    def __call__(self, other):
        pp = []
        for people in other:
            ppl.name = people[0]
            ppl.sex = people[1]
            ppl.age = people[2]
            ppl.addr = people[3]
            pp.append(ppl)
        return pp 
    
```
